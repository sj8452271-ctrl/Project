```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import warnings
warnings.filterwarnings('ignore')

print("All libraries imported successfully!")
```

```python
df = pd.read_csv('zomato.csv', index_col=0)

# Drop the duplicate unnamed index column
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# Check for missing values
print("Missing Values:")
print(df.isnull().sum())

# Drop rows with missing values
df.dropna(inplace=True)

# Rename columns for easier use
df.columns = ['restaurant_name', 'restaurant_type', 'rating',
               'num_ratings', 'avg_cost', 'online_order',
               'table_booking', 'cuisines', 'area', 'local_address']

# Convert rating and cost to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['avg_cost'] = pd.to_numeric(df['avg_cost'], errors='coerce')
df.dropna(inplace=True)

print("\nCleaned Dataset Shape:", df.shape)
print("\nData Types:")
print(df.dtypes)
df.head()
```

```python
print("Basic Statistics:")
print(df[['rating', 'num_ratings', 'avg_cost']].describe())

print("\nOnline Order Distribution:")
print(df['online_order'].value_counts())

print("\nTable Booking Distribution:")
print(df['table_booking'].value_counts())

print("\nTop 5 Areas with Most Restaurants:")
print(df['area'].value_counts().head())

print("\nTop 5 Restaurant Types:")
print(df['restaurant_type'].value_counts().head())
```

```python
fig, axes = plt.subplots(3, 2, figsize=(16, 18))
fig.suptitle('Zomato Food Delivery - Business Insights Dashboard',
             fontsize=18, fontweight='bold', y=1.01)

# ── Chart 1: Top 10 Most Popular Areas ──
top_areas = df['area'].value_counts().head(10)
axes[0, 0].barh(top_areas.index, top_areas.values, color='steelblue')
axes[0, 0].set_title('Top 10 Areas by Number of Restaurants', fontweight='bold')
axes[0, 0].set_xlabel('Number of Restaurants')
axes[0, 0].invert_yaxis()

# ── Chart 2: Online Order vs Table Booking ──
online_counts = df['online_order'].value_counts()
table_counts = df['table_booking'].value_counts()
x = ['Yes', 'No']
axes[0, 1].bar(x, [online_counts.get('Yes', 0), online_counts.get('No', 0)],
               color=['green', 'red'], alpha=0.7, label='Online Order', width=0.4, align='center')
axes[0, 1].bar([i + 0.4 for i in range(2)],
               [table_counts.get('Yes', 0), table_counts.get('No', 0)],
               color=['blue', 'orange'], alpha=0.7, label='Table Booking', width=0.4)
axes[0, 1].set_title('Online Order vs Table Booking Availability', fontweight='bold')
axes[0, 1].set_xlabel('Available?')
axes[0, 1].set_ylabel('Number of Restaurants')
axes[0, 1].legend()
axes[0, 1].set_xticks([0.2, 1.2])
axes[0, 1].set_xticklabels(['Yes', 'No'])

# ── Chart 3: Rating Distribution ──
axes[1, 0].hist(df['rating'], bins=20, color='coral', edgecolor='black')
axes[1, 0].set_title('Distribution of Restaurant Ratings', fontweight='bold')
axes[1, 0].set_xlabel('Rating (out of 5)')
axes[1, 0].set_ylabel('Number of Restaurants')
axes[1, 0].axvline(df['rating'].mean(), color='blue', linestyle='--',
                    label=f'Mean: {df["rating"].mean():.2f}')
axes[1, 0].legend()

# ── Chart 4: Top 10 Most Popular Cuisines ──
all_cuisines = df['cuisines'].str.split(',').explode().str.strip()
top_cuisines = all_cuisines.value_counts().head(10)
axes[1, 1].bar(top_cuisines.index, top_cuisines.values, color='mediumpurple')
axes[1, 1].set_title('Top 10 Most Popular Cuisines', fontweight='bold')
axes[1, 1].set_xlabel('Cuisine Type')
axes[1, 1].set_ylabel('Count')
axes[1, 1].tick_params(axis='x', rotation=45)

# ── Chart 5: Average Cost Distribution ──
axes[2, 0].hist(df['avg_cost'], bins=25, color='teal', edgecolor='black')
axes[2, 0].set_title('Average Cost for Two People', fontweight='bold')
axes[2, 0].set_xlabel('Cost (₹)')
axes[2, 0].set_ylabel('Number of Restaurants')
axes[2, 0].axvline(df['avg_cost'].mean(), color='red', linestyle='--',
                    label=f'Mean: ₹{df["avg_cost"].mean():.0f}')
axes[2, 0].legend()

# ── Chart 6: Average Rating by Restaurant Type ──
avg_rating_type = df.groupby('restaurant_type')['rating'].mean().sort_values(ascending=False).head(8)
axes[2, 1].bar(avg_rating_type.index, avg_rating_type.values, color='goldenrod')
axes[2, 1].set_title('Average Rating by Restaurant Type (Top 8)', fontweight='bold')
axes[2, 1].set_xlabel('Restaurant Type')
axes[2, 1].set_ylabel('Average Rating')
axes[2, 1].tick_params(axis='x', rotation=45)
axes[2, 1].set_ylim(0, 5)

plt.tight_layout()
plt.savefig('zomato_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
print("Dashboard saved as 'zomato_dashboard.png'")
```

```python
print("=" * 55)
print("       KEY BUSINESS INSIGHTS SUMMARY")
print("=" * 55)

print(f"\n📊 Total Restaurants Analyzed : {len(df)}")
print(f"⭐ Average Rating             : {df['rating'].mean():.2f} / 5")
print(f"💰 Average Cost for Two       : ₹{df['avg_cost'].mean():.0f}")
print(f"📍 Total Areas Covered        : {df['area'].nunique()}")

print(f"\n🏆 Most Popular Area          : {df['area'].value_counts().index[0]}")
print(f"🍽️  Most Common Restaurant Type: {df['restaurant_type'].value_counts().index[0]}")
print(f"🥘 Most Popular Cuisine       : {all_cuisines.value_counts().index[0]}")

online_pct = (df['online_order'] == 'Yes').mean() * 100
table_pct  = (df['table_booking'] == 'Yes').mean() * 100
print(f"\n📱 Restaurants with Online Order : {online_pct:.1f}%")
print(f"🪑 Restaurants with Table Booking: {table_pct:.1f}%")

top5 = df.groupby('restaurant_name')['num_ratings'].sum().sort_values(ascending=False).head(5)
print(f"\n🌟 Top 5 Most Rated Restaurants:")
for i, (name, count) in enumerate(top5.items(), 1):
    print(f"   {i}. {name} — {int(count)} ratings")

print("=" * 55)
```
